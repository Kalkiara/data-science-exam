```python
# load packages

import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
```

```python
# load data

df = pd.read_csv("/work/MiaJacobsen#3200/data-science-exam/data/top_songs_by_month.csv")
```

```python
# inspect
print(df.shape)
print(df.dtypes)
df.head()
```

```python
# making date datetime
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
# looking at a subset of the data
df1 = df.iloc[0:2000, :]
df1 = df1[['date', 'danceability']]
```

```python
# trying to smooth the line (medium success)
df1['mov_avg'] = df1['danceability'].rolling(365*3).mean()
```

```python
# exploratory plotting
sns.lineplot(data = df1, x = "date", y = "mov_avg")
plt.xticks(rotation = 45)
plt.show()
```
